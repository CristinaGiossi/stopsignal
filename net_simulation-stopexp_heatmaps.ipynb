### Varying SS amplitudes injected to STN and D2 - Loop and heatmaps

```python
stop_percentage = []
ampl_valuesSTN = []
ampl_valuesD2 = []
onset_values = [] #Same onset value for both stop signals 
duration_values = [] #Same duration value for both stop signals 

GPeP_sum_singletrials = []
GPeP_Fi_singletrials = []
GPeP_FR_mean_tot = []

GPeA_sum_singletrials = []
GPeA_Fi_singletrials = []
GPeA_FR_mean_tot = []

for amplitudeSTN in np.arange(0.25, 0.85, 0.1):
    
    for amplitudeD2 in np.arange(0.5, 1.1, 0.1):
        
        
        environment = {
                'params': None, #neuron parameters (init_params.py)
                'pops': None,   #population parameters (init_params.py)
                'receps' : None, #receptor parameters (init_params.py)
                'base' : None, #baseline stimulation parameters (init_params.py)
                'dpmns' : None, #dopamine related parameters (init_params.py)
                'd1' : None, #D1-MSNs population related parameters (init_params.py)
                'd2' : None, #D2-MSNs population related parameters (init_params.py)
                'channels' : pd.DataFrame([["left"],["right"]], columns=['action']), #action channels related parameters (init_params.py)
                #'actionchannels' : pd.DataFrame([[1],[2]], columns=['action']), #labels for the actions (init_params.py)
                #'actionchannels' : pd.DataFrame([["left"],["right"]], columns=['action']), #labels for the actions (init_params.py)
                'newpathways' : None, #connectivity parameters (popconstruct.py)
                'Q_support_params': None, #initialization of Q-values update (qvalues.py) 
                'Q_df':None, #initialized Q-values df (qvalues.py)  
                'n_trials': 100, #number of trials (generateepochs.py)
                'volatility': 1, #frequency of changepoints (generateepochs.py)
                'conflict': 0.7, #probability of the preferred choice (generateepochs.py)
                'reward_mu': 1, #mean for the magnitude of the reward (generateepochs.py)
                'reward_std': 0.1, #std for the magnitude of the reward (generateepochs.py)
                'maxstim': 0.95, # amplitude of the cortical input over base line #0.85
                'sustainedfraction': 0.75,

                #STN
                'stop_signal_present': True,
                'stop_signal_probability': 1., #probability of trials that will get the stop signal / list of trial numbers
                'stop_signal_amplitude': amplitudeSTN, # amplitude of the stop signal over base line
                'stop_signal_onset': 55., #in ms #antes 200
                'stop_signal_duration' : 160., #in ms
                'stop_signal_channel': "all", #"all" (all channels are given the stop signal) 
                                            #/"any" (channel given the stop signal is chosen randomly)
                                            # [list of channels] == subset of channels given the stop signa

                #D2STR
                'stop_signal_present_2': True,
                'stop_signal_probability_2': 1., #probability of trials that will get the stop signal / list of trial numbers
                'stop_signal_amplitude_2': amplitudeD2, # amplitude of the stop signal over base line
                'stop_signal_onset_2': 55., #in ms #antes 200
                'stop_signal_duration_2' : 160., #in ms
                'stop_signal_channel_2': "all", #"all" (all channels are given the stop signal) 
                                            #/"any" (channel given the stop signal is chosen randomly)
                                            # [list of channels] == subset of channels given the stop signal

                'decision_channel' : '1', #'all'(all channels)
                                          # '0' : left channel considered 
                                          # '1' : right channel considered
        }                  
        results = cbgt.ExecutionManager(cores=7).run([pl]*1,[environment]*1)

        datatables = cbgt.collateVariable(results,'datatables')
        plt_func.rename_columns(results[0])
        popfreqs = cbgt.collateVariable(results,'popfreqs')

        stop_number = 0
        for i in datatables[0].index:
            if datatables[0].iloc[i]['decision'] == 'stop':
                stop_number += 1
        stop_percentage.append(stop_number/results[0]['n_trials'])
        ampl_valuesSTN.append(amplitudeSTN)
        ampl_valuesD2.append(amplitudeD2)
        onset_values.append(results[0]['stop_signal_onset'])
        
        for i in datatables[0].index:
    
            GPeP_sum_singletrials.append(sum(popfreqs[0]['GPeP_right'][datatables[0].stimulusstarttime[i]+int(results[0]['stop_signal_onset']):
                 datatables[0].stimulusstarttime[i]+int(results[0]['stop_signal_onset'])+int(results[0]['stop_signal_duration'])]))
            GPeP_Fi_singletrials.append(GPeP_sum_singletrials[i]/results[0]['stop_signal_duration'])
            
            GPeA_sum_singletrials.append(sum(popfreqs[0]['GPeA_right'][datatables[0].stimulusstarttime[i]+int(results[0]['stop_signal_onset']):
                 datatables[0].stimulusstarttime[i]+int(results[0]['stop_signal_onset'])+int(results[0]['stop_signal_duration'])]))
            GPeA_Fi_singletrials.append(GPeA_sum_singletrials[i]/results[0]['stop_signal_duration'])
        
        
        GPeP_FR_mean_tot.append(sum(GPeP_Fi_singletrials)/results[0]['n_trials'])
        GPeA_FR_mean_tot.append(sum(GPeA_Fi_singletrials)/results[0]['n_trials'])
                                
        cbgt.saveResults(results, data_dir+'stop_data_heatmap_STNampl'+str(amplitudeSTN)+'_D2ampl_'+str(amplitudeD2)+'seed'+str(seed),['popfreqs','popdata', 'datatables'])
    
```

```python
saveresults_vars(stop_percentage, data_dir+'/heatmap_loop/loop_heatmap_stopprobs'+'_seed'+str(seed))
saveresults_vars(onset_values, data_dir+'/heatmap_loop/loop_heatmap_onsetvalues'+'_seed'+str(seed))
saveresults_vars(ampl_valuesSTN, data_dir+'/heatmap_loop/loop_heatmap_amplvalues_STN'+'_seed'+str(seed))
saveresults_vars(ampl_valuesD2, data_dir+'/heatmap_loop/loop_heatmap_amplvalues_D2'+'_seed'+str(seed))

saveresults_vars(GPeP_Fi_singletrials, data_dir+'/heatmap_loop/loop_heatmap_GPeP_Fi_singletrials'+'_seed'+str(seed))
saveresults_vars(GPeP_FR_mean_tot, data_dir+'/heatmap_loop/loop_heatmap_GPeP_FR_mean_tot'+'_seed'+str(seed))

saveresults_vars(GPeA_Fi_singletrials, data_dir+'/heatmap_loop/loop_heatmap_GPeA_Fi_singletrials'+'_seed'+str(seed))
saveresults_vars(GPeA_FR_mean_tot, data_dir+'/heatmap_loop/loop_heatmap_GPeA_FR_mean_tot'+'_seed'+str(seed))
```
